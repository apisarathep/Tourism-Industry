```python
url= "https://drive.google.com/file/d/1thFGb4xbh7Y4cshPPSiqLm-FJwnVaf5m/view?usp=sharing"
file_id = url.split('/')[-2]
dwn_url='https://drive.google.com/uc?id=' + file_id

df = pd.read_csv(dwn_url)
data = pd.DataFrame(df)
df
```

```python
# Set the style of the plot
plt.style.use('ggplot')

# Create a figure object and set the size
fig = plt.figure(figsize=(10, 8))

# Create the data for the plot
data_Female = (data['เพศของคุณคือเพศอะไร'] == 'หญิง').sum()
data_Man = (data['เพศของคุณคือเพศอะไร'] == 'ชาย').sum()
data_LGBT = (data['เพศของคุณคือเพศอะไร'] == 'LGBTQ+').sum()
data_None = (data['เพศของคุณคือเพศอะไร'] == 'ไม่ต้องการระบุ').sum()
genterIs = data_Female + data_Man + data_LGBT + data_LGBT

# Create the height and label for the bars
height = [data_Female, data_Man, data_LGBT, data_None]
bars = ['ผู้หญิง', 'ผู้ชาย', 'LGBTQ+', 'ไม่ต้องการระบุเพศ']

# Set the title and labels for the plot
plt.title("กราฟแสดงเพศของนักศึกษามหาวิทยาลัยธรรมศาสตร์",  pad=30, fontsize=16)
plt.xlabel("เพศ", labelpad=20, fontsize=14)
plt.ylabel("จำนวนผู้ตอบแบบสอบถาม", labelpad=20,fontsize=14)

# Create the bar plot
plt.bar(bars, height, width=0.5, color=['#D0F4DE', '#C0E4F6', '#FFA1D9', '#E8CFF8'])

# Set the background color
plt.gca().set_facecolor('#F8FDFF')

# Add index and percent value for each bar
for i, h in enumerate(height):
    plt.text(i, h+0.5, f"{h:,}\n({h/genterIs:.1%})", ha='center', fontsize=12)


# Show the plot
plt.show()
```

```python
firstID = (data['คุณมีเลขนักศึกษา 2 ตัวเเรกคืออะไร\n\n(เช่น รหัสนักศึกษาคือ 6409650700 สองตัวแรกคือ 64)'] == 62).sum()
secondID = (data['คุณมีเลขนักศึกษา 2 ตัวเเรกคืออะไร\n\n(เช่น รหัสนักศึกษาคือ 6409650700 สองตัวแรกคือ 64)'] == 63).sum()
thirdID = (data['คุณมีเลขนักศึกษา 2 ตัวเเรกคืออะไร\n\n(เช่น รหัสนักศึกษาคือ 6409650700 สองตัวแรกคือ 64)'] == 64).sum()
fourthID = (data['คุณมีเลขนักศึกษา 2 ตัวเเรกคืออะไร\n\n(เช่น รหัสนักศึกษาคือ 6409650700 สองตัวแรกคือ 64)'] == 65).sum()

labels = ['รหัส 62', 'รหัส 63', 'รหัส 64', 'รหัส 65']
sizes = [firstID, secondID, thirdID, fourthID]
colors = ["#D0F4DE", "#C0E4F6", "#FFBBDA", "#E8CFF8"]
total = sum(sizes)

plt.pie(sizes, labels=labels, colors=colors, autopct=lambda pct: "{:.1f}%\n({:.0f})".format(pct, pct*total/100),textprops={'fontsize': 10, 'color': 'black'})
plt.axis('equal')
plt.title('กราฟแสดงรหัสนักศึกษาสองตัวแรก(คน)', pad=22, fontsize=16)

# Move legend outside the pie chart
plt.legend(labels, bbox_to_anchor=(1.28, 0.8), loc="center right")

plt.show()
```
